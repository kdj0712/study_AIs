```python
import pandas as pd
import pyLDAvis
import pyLDAvis.lda_model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.datasets import fetch_20newsgroups
# 데이터 로드
data = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))
documents = data.data
```

```python
df_120text = pd.read_csv("https://raw.githubusercontent.com/corazzon/web_scraping/master/seoul-120/seoul-120-text.csv")
df_120text[:3]
```

```python
tfidfVectorizer = TfidfVectorizer(stop_words='english', max_df=0.95 ,min_df=2)
```

```python
tfidf_documents = tfidfVectorizer.fit_transform(documents)
tfidf_documents.toarray()[:1]
```

```python
tfidf_documents.toarray().shape
```

```python
import numpy as np
np.where(tfidf_documents.toarray()[:1]>0)
```

```python
latentDirichletAllocation_3 = LatentDirichletAllocation(n_components=3, n_jobs=-1)
```

```python
latentDirichletAllocation_5 = LatentDirichletAllocation(n_components=5, n_jobs=-1)
```

```python
latentDirichletAllocation_7= LatentDirichletAllocation(n_components=7, n_jobs=-1)
```

```python
LDA3 = latentDirichletAllocation_3.fit(tfidf_documents)
LDA5 = latentDirichletAllocation_5.fit(tfidf_documents)
LDA7 = latentDirichletAllocation_7.fit(tfidf_documents)
```

```python
vis3 = pyLDAvis.lda_model.prepare(LDA3, tfidf_documents, tfidfVectorizer)
vis5 = pyLDAvis.lda_model.prepare(LDA5, tfidf_documents, tfidfVectorizer)
vis7 = pyLDAvis.lda_model.prepare(LDA7, tfidf_documents, tfidfVectorizer)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.display(vis3)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.display(vis5)
```
